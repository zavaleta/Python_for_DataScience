## 1. Exemplos usando numpy

- *ndarray*: Vetor homogêneo multidimensional que possui uma tupla de índices. 

```python
# importando a biblioteca numpy
import numpy as np
```

```python
pesos = np.array([50.,55.5,53.6,60.7,70.9,56.9])
pesos
```

```python
# media aritmetica
media = pesos.mean()
print("Média Aritmética =",media)
# máximo dos pesos
maximo = pesos.max()
print("Máxino dos pesos =",maximo)
# mínimo dos pesos
minimo = pesos.min()
print("Mínimo dos pesos =",minimo)
```

```python
# função linspace
weights = np.linspace(50,60,num=20)
print('Pesos =',weights)
```

```python
weights1 = np.linspace(1,50,num=20)
print('Pesos =',weights1)
```

```python
# diferença de vetores ndarray
dif = weights- weights1
print("pesos =",dif)
```

```python
# carregando os dados do dataset
dados = load_iris()
caracteristicas = dados['data']
nomes_das_caracteristicas = dados['feature_names']
objetivo = dados['target']
```

```python
# graficar as classes com cores diferentes
for t,marker,c in zip(range(3),">ox","rgb"):
    plt.scatter(caracteristicas[objetivo == t,0],caracteristicas[objetivo == t,1],marker=marker,c=c)
    #plt.xlim([4,8])
    #plt.ylim([2,4])
    
```

```python
# usando PCA
from sklearn.decomposition import PCA
```

```python
# usando o PCA
pca_X = dados.data
pca_y = dados.target
pca = PCA(n_components=2)
# redução de dimensões
reduced_X = pca.fit_transform(pca_X)
```

```python
red_x, red_y = [], []
blue_x, blue_y = [], []
green_x, green_y = [], []
for i in range(len(reduced_X)):
    if pca_y[i] == 0:
        red_x.append(reduced_X[i][0])
        red_y.append(reduced_X[i][1])
    elif pca_y[i] == 1:
        blue_x.append(reduced_X[i][0])
        blue_y.append(reduced_X[i][1])
    else:
        green_x.append(reduced_X[i][0])
        green_y.append(reduced_X[i][1])
        #
plt.scatter(red_x, red_y, c='r', marker='+',label='Classe 0')
plt.scatter(blue_x, blue_y, c='b', marker='D',label='Classe 1')
plt.scatter(green_x, green_y, c='g', marker='o',label='Classe 2')
plt.title('Usando PCA no dataset Iris')
plt.legend()
plt.show()
```
